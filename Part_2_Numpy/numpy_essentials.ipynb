## Задача 4

Найти решение для дифуравнения $\left(\frac{d^2}{dx^2} + (10x)^2\right) f = \lambda f$ с граничными условиями $f(0)=f(1)=0$

* используя метод собственных векторов и уравнение $\frac{d^2 f}{dx^2} \approx \frac{f_{i+1} + f_{i-1} - 2f_i}{\Delta x^2}$ где $f_i = f(i \Delta x)$ и $i$ это целое число. 

```python
N = 1000
x = np.linspace(0,1,N+1)
dx = x[1]-x[0]

main_diag = -2*np.ones(N-1)
off_diag = np.ones(N-2)
derivative_matrix = (np.diag(main_diag) + np.diag(off_diag, k=1)+ np.diag(off_diag, k=-1))/dx**2
x2_matrix = np.diag((10*x[1:-1])**2)
LHS_matrix = derivative_matrix #+ x2_matrix 
```

```python
w, v = np.linalg.eigh(LHS_matrix)
```

```python
plt.plot(-1*v[:,-1])
plt.plot(-1*v[:,-2])
plt.plot(-1*v[:,-3])
plt.plot(-1*v[:,-4])
```
